```python
import pandas as pd
import numpy as np
from herramientas import obtener_resumen_valores_unicos
```

```python
df_train = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')
```

```python
data_to_add = {
    'outcome': [8] * len(df_test),

}

# Convierte el diccionario en un nuevo DataFrame
new_columns_df = pd.DataFrame(data_to_add)

# Concatena el DataFrame original ('df') con el nuevo DataFrame
df_test = pd.concat([df_test, new_columns_df], axis=1)
```

```python
df= pd.concat([df_test, df_train], axis=0)
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# Obtén una lista de las columnas categóricas (tipos object)
columnas_categoricas = df.select_dtypes(include=['object']).columns

# Itera a través de las columnas categóricas y muestra los valores únicos
for columna in columnas_categoricas:
    valores_unicos = df[columna].unique()
    print(f'Valores únicos en la columna "{columna}":')
    for valor in valores_unicos:
        print(valor)
    print('\n')
```

```python
import pandas as pd

# Agrupar por la columna 'temp_of_extremities' y calcular las estadísticas
grouped = df_test.groupby('temp_of_extremities')['rectal_temp']

# Calcular los valores deseados
min_temp = grouped.min()
max_temp = grouped.max()
mean_temp = grouped.mean()
median_temp = grouped.median()

# Mostrar los resultados
result_df = pd.DataFrame({
    'Mínimo': min_temp,
    'Máximo': max_temp,
    'Media': mean_temp,
    'Mediana': median_temp
})

print(result_df)
```
